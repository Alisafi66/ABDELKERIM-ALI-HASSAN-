```python
# Handling outliers.
def visualize_outliers(data, target_c): # This function that takes two parameters: data (a DataFrame) and target_c (the target/dependent column to drop from the visualization).
    data = data.drop(target_c, axis=1) # This line drops the specified target column (target_c) from the DataFrame (data). This is done to avoid visualizing the boxplot for the target column itself.
    for c in data.columns: # Inside the loop, it checks if the data type of the column is either 'int' or 'float'. If it is, a boxplot is created using sns.boxplot(data[c]).
        if (data[c].dtype == 'int') or (data[c].dtype == 'float'):
            sns.boxplot(data[c]);
            plt.title(f'Column: {c}')
            plt.show()
            
visualize_outliers(data=df, target_c='BodyFat')
```

```python
def object_feature(data):
    """
    Visualize the distribution of categorical features in a DataFrame.

    Parameters:
    - data: DataFrame
    """
    for c in data.columns:
        # Check if the column is of object (string) type
        if data[c].dtype == 'object':
            # Plot the value counts as a bar plot
            data[c].value_counts().plot(kind='bar')
            plt.title(f'Column: {c}');
            plt.xlabel('');
            plt.show()
            
object_feature(df)
```

```python
# majority of the features in the dataset are continuous numerical data
# define a helper function which plots a histogram and boxplot for visualizing the distribution of continuous data

def viz_cont(feat, feat_name):
    figure, axes = plt.subplots(1, 2, figsize=(15, 5), gridspec_kw={'width_ratios':[2, 1]})
    sns.histplot(df[feat], ax=axes[0])
    axes[0].set_title(f'Histogram: Distribution of {feat_name}')
    sns.boxplot(df[feat], ax=axes[1])
    axes[1].set_title(f'Boxplot: Central Tendency of {feat_name}')
    plt.tight_layout()
    plt.show()
    
    # calculate interquartile range 
    Q3, Q1 = np.percentile(df[feat], [75, 25])
    IQR = Q3 - Q1

    # define the upper and lower bound
    upper_bound = Q3 + IQR
    lower_bound = Q1 - IQR

    # subset the instances that are potentially outliers
    outliers = df[(df[feat] > upper_bound) | (df[feat] < lower_bound)]
    outliers_percent = round((len(outliers) / len(df)) * 100, 2)
    print(f'There are {len(outliers)} instances, representing approximately {outliers_percent}% potential outliers of {feat_name} within the dataset.')
    display(outliers)
```

```python
# transform the feature matrix for train and test data
Xtrain_transformed = featureTransformer.fit_transform(Xtrain)
Xtest_transformed = featureTransformer.transform(Xtest)
```

```python
# define a helper function to evaluate model performance based on MAE, MSE and adjusted R-squared
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

def model_eval(X,y,clf,colTag):
    ypred = clf.predict(X)
    mae = mean_absolute_error(y,ypred)
    mse = mean_squared_error(y,ypred)
    rmse = np.sqrt(mse)
    
    # sklearn doesn't have a function to calculate the adjusted R2-squared
    # to perform a manual calculation based on R-squared value
    
    r2 = r2_score(y,ypred)
    
    # get the number of observations and features
    n = len(y)
    k = X.shape[1]
    
    # calculate adjusted R-squared
    adj_R2 = 1 - ((1 - r2) * (n - 1) / (n - k - 1))
    
    # sklearn doesn't have a function to calculate the MAPE
    # calculate absolute percentage error for each data point
    ape = np.abs((y - ypred) / y)

    # remove NaN or Inf values (in case of zeros in y_actual)
    ape = ape[~np.isnan(ape) & ~np.isinf(ape)]

    # calculate mean absolute percentage error (MAPE)
    mape = np.mean(ape) 
    
    
    metrics = ['MAE','MSE','RMSE','MAPE','Adjusted R-Squared']
    values = [mae,mse,rmse,mape,adj_R2]
    
    res = pd.DataFrame(values, columns=[colTag], index=metrics)
    return res
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error

# create and train the linear regression model
linear_regressor = LinearRegression()
linear_regressor.fit(Xtrain_transformed, ytrain)

# model performance on train data
linear_regression_train = model_eval(Xtrain_transformed, ytrain, linear_regressor,'Linear_train')

# model performance on unseen test data
linear_regression_test = model_eval(Xtest_transformed, ytest, linear_regressor,'Linear_test')

# model performance on train and test datasets
performance = pd.concat([linear_regression_train,linear_regression_test], axis=1)
display(performance)
```
